```python
def consecutive_trues(snip):
    #count number of consecutives from back
    cnt=0
    for i in reversed(xrange(1,snip.size)):
        #print '%d %f' % (i,snip[i])
        if bool(snip[i]) and bool(snip[i-1]):
            cnt+=1
        else:
            break
            
    return cnt

```

```python
s=bdu.initBBGSession()
dxy=bdu.getTimeSeries(s,'DXY Index','PX_LAST',19710101)
cpi=bdu.getTimeSeries(s,'CPI YOY Index','PX_LAST',19710101)
gdp=bdu.getTimeSeries(s,'GDP CQOQ Index','PX_LAST',19710101)

dxy.resample('W-FRI','last').pct_change() > 0
```

```python
rslt=pd.rolling_apply(dxy.resample('W-FRI','last').pct_change()>0,15,consecutive_trues).dropna()
rslt.plot(title='DXY continuous weekly increases vs. GDP')
gdp.plot(figsize=(12,10))
```
